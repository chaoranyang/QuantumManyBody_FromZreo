## 2.一维相互作用定性分析
$\textcolor{lime}{完全非局域化}$

原子均匀分布在整个系统中，原子之间的相互作用非常充分。而因为它们之间的相互作用为短程排斥相互作用，每个原子的独立运动受到较强的限制。每个原子的平均能量 $e_{1} = E / N$ ，认为动能忽略不计，只有相互作用能

$$
e_{1}\simeq \frac{1}{2} gn
$$

其中，$g$ 是原子相互作用耦合常数，定义与前面章节相同；$n$ 为原子数密度

$\textcolor{lime}{完全局域化}$

每个原子几乎可以在体积为 $1 / n = V / N$ 的空间中独立自由运动，原子之间距离过远，感受不到相互作用，每个原子能量 $e_{2}$ 可以忽略相互作用能，只有动能

$$
e_{2}\simeq \frac{\hbar^{2}}{2m a^{2}\Delta x^{2}} = \frac{\hbar^{2}n^{2 / d}}{m}
$$

其中，$d$ 是原子所在空间的维度，而 $\Delta x = 1 / n^{d}$ 是每个原子在每一个维度上的近似的自由运动空间。

$\textcolor{lime}{强相互作用系统}$

 $e_{1} \gg e_{2}$，即原子的相互作用远大于其自由运动。在三维空间 $(d = 3)$ 中，这也就意味着

$$
n^{1 / 3} \gg \frac{\hbar^{2}}{m g}
$$

即原子数密度较高的系统其相互作用较强，这也符合 直觉。而相反，对于一维系统 $(d = 1)$ 而言，我们会得到

$$
n^{-1} \gg \frac{\hbar^{2}}{m g}
$$

而这个式子告诉我们，$\textcolor{yellow}{原子数密度越低的一维系统越处于强相互作用}$，这和我们的直觉是完全相反！对于一维玻色子，当它处于极强相互作用和低温，会被费米化，而其部分性质和理想费米子完全一致。

定义无量纲的参数

$$\textcolor{red}{\gamma = \frac{m g}{\hbar^{2}n}}
$$

这个参数也叫做$\textcolor{red}{Lieb-Liniger 参数}$。$\gamma\gg1$强相互作用极限，$\gamma\ll 1$强相互作用极限


# 3.Lieb-Liniger模型与Bethe拟设
$\textcolor{lime}{主要问题}$

$\textcolor{lime}{1、}$绕开多体散射，直接从二体薛定谔方程出发推导

$\textcolor{lime}{2、}$"Bethe拟设本质是把相互作用转化为周期边界条件”的含义：用周期边界条件在$A(\mathcal{P})$的循环关系中消去$A(\mathcal{P})$，得到单纯关于$k$的方程

Lieb-Liniger模型
$$\textcolor{yellow}{H = -\sum_{i=1}^N \frac{\partial^2}{\partial x_i^2} + g \sum_{i<j}^N \delta(x_i - x_j)}$$

Bethe拟设
$$\textcolor{yellow}{
\psi(x_1, \cdots, x_N) = \sum_{\mathcal{P}, \mathcal{Q}} A_\mathcal{P} e^{i \sum_{j=1}^N k_{\mathcal{P}_j} x_{\mathcal{Q}_j}} \theta(x_{\mathcal{Q}_1} < \cdots < x_{\mathcal{Q}_N}).}
$$
其中$\mathcal{P}$和$\mathcal{Q}$是对N个粒子的排列



$\textcolor{lime}{两粒子情况}$
$$\psi(x_1, x_2) = \theta(x_2 - x_1)[A_{12}e^{i(k_1x_1 + k_2x_2)} + A_{21}e^{i(k_2x_1 + k_1x_2)}]
+ \theta(x_1 - x_2)[A_{12}e^{i(k_1x_2 + k_2x_1)} + A_{21}e^{i(k_2x_2 + k_1x_1)}], $$

其中，$ A_{12} $ 和 $ A_{21} $ 与 $ k_1 $, $ k_2 $ 有关；$ \theta(x) $ 是阶梯函数，当 $ x > 0 $ 时，$ \theta(x) = 1 $，当 $ x < 0 $ 时，$ \theta(x) = 0 $。利用质心坐标，令
$
X = \frac{x_1 + x_2}{2}, \quad Y = x_2 - x_1,
$
可以得到
$
x_1 = X - \frac{1}{2}Y, \quad x_2 = X + \frac{1}{2}Y,
$
求导，得
$$
\frac{\partial^{2}}{\partial x_{1}^{2}} +\frac{\partial^{2}}{\partial x_{2}^{2}} = \frac{1}{2}\frac{\partial^{2}}{\partial X^{2}} +2\frac{\partial^{2}}{\partial Y^{2}}.
$$
由薛定谔方程 $H\psi = E\psi$ 可得到

$$
\left[-\frac{1}{2}\frac{\partial^{2}}{\partial X^{2}} -2\frac{\partial^{2}}{\partial Y^{2}}\right]\psi (X,Y) + g\delta (Y)\psi (X,Y) = E\psi (X,Y).
$$

对上式作积分

$$
\int_{-\epsilon}^{+\epsilon}\frac{\partial^{2}\psi}{\partial Y^{2}}\mathrm{d}Y = \int_{-\epsilon}^{+\epsilon}\frac{g}{2}\delta (Y)\psi \mathrm{d}Y - \frac{1}{2}\int_{-\epsilon}^{+\epsilon}E\psi \mathrm{d}Y - \frac{1}{4}\int_{-\epsilon}^{+\epsilon}\frac{\partial^{2}\psi}{\partial X^{2}}\mathrm{d}Y(\epsilon \to 0)
$$

得

$$
\frac{\partial\psi}{\partial Y}\big|_{Y = 0^{+}} - \frac{\partial\psi}{\partial Y}\big|_{Y = 0^{-}} = \frac{g}{2}\psi \big|_{Y = 0}.
$$

又因为 $\frac{\partial}{\partial Y} = \frac{\partial}{\partial x_{1}}\frac{\partial x_{1}}{\partial Y} +\frac{\partial}{\partial x_{2}}\frac{\partial x_{2}}{\partial Y} = -\frac{1}{2}\frac{\partial}{\partial x_{1}} +\frac{1}{2}\frac{\partial}{\partial x_{2}}$ 取 $\epsilon \to 0$ 的极限,可写为

$$
\left(\frac{\partial}{\partial x_{2}} -\frac{\partial}{\partial x_{1}}\right)\psi \big|_{x_{2} = x_{1}^{+}} - \left(\frac{\partial}{\partial x_{2}} -\frac{\partial}{\partial x_{1}}\right)\psi \big|_{x_{2} = x_{1}^{-}} = g\psi \big|_{x_{2} = x_{1}}.
$$

由于玻色子满足交换对称性 $\left(\frac{\partial}{\partial x_{2}} -\frac{\partial}{\partial x_{1}}\right)\psi \big|_{x_{2} = x_{1}^{+}} = \left(\frac{\partial}{\partial x_{1}} -\frac{\partial}{\partial x_{2}}\right)\psi \big|_{x_{2} = x_{1}^{-}}$ ,所以

$$
\left(\frac{\partial}{\partial x_{2}} -\frac{\partial}{\partial x_{1}}\right)\psi \big|_{x_{2} = x_{1}^{+}} = c\psi \big|_{x_{2} = x_{1}}
$$

令$c=g/2$

$$
\frac{A_{12}}{A_{21}}=\frac{\mathrm{i}k_{2}-\mathrm{i}k_{1}+c}{k_{2}-\mathrm{i}k_{1}-c}=-\frac{c+\mathrm{i}(k_{2}-k_{1})}{c-\mathrm{i}(k_{2}-k_{1})}=-\mathrm{e}^{\mathrm{i}\theta(k_{2}-k_{1})}
$$

这里, $\theta (x) = 2\arctan (x / c)$

$\textcolor{lime}{N个粒子系统}$

$$  
\left[-\sum_{i = 1}^{N}\frac{\partial^{2}}{\partial x_{i}^{2}} +g\sum_{i< j}^{N}\delta (x_{i} - x_{j})\right]\psi = E\psi ,
$$

利用质心坐标  

$$  
\begin{array}{l}{{X=\frac{x_{j}+x_{k}}{2},\quad Y=x_{j}-x_{k},}}\\ {{x_{j}=X-\frac{1}{2}Y,\quad x_{k}=X+\frac{1}{2}Y.}}\end{array}
$$

同样我们可以得到  

$$  
\frac{\partial^{2}}{\partial x_{j}^{2}} +\frac{\partial^{2}}{\partial x_{k}^{2}} = \frac{1}{2}\frac{\partial^{2}}{\partial X^{2}} +2\frac{\partial^{2}}{\partial Y^{2}}.
$$

由薛定谔方程进一步可得  

$$  
\left(-\frac{\partial^{2}}{\partial x_{1}^{2}} -\dots -\frac{1}{2}\frac{\partial^{2}}{\partial X^{2}} -2\frac{\partial^{2}}{\partial Y^{2}} -\dots -\frac{\partial^{2}}{\partial x_{N}^{2}}\right)\psi +\dots +g\delta (x_{j} - x_{k})\psi +\dots = E\psi .
$$

对 $Y $作无穷小积分  

$$  
\left(-\frac{\partial^{2}}{\partial x_{1}^{2}} -\dots -\frac{1}{2}\frac{\partial^{2}}{\partial X^{2}} -\frac{\partial^{2}}{\partial x_{N}^{2}}\right)\psi (Y)|_{-\epsilon}^{+\epsilon} - 2\frac{\partial}{\partial Y}\psi |_{-\epsilon}^{+\epsilon} + g\psi |_{Y = 0} = E\psi (Y)|_{-\epsilon}^{+\epsilon}
$$

如上述两个粒子系统处理方法一样, $N$ 粒子系统满足的边界条件为  

$$  
\left(\frac{\partial}{\partial x_{j}} -\frac{\partial}{\partial x_{k}}\right)\psi \big|_{x_{j} = x_{k}^{+}} - \left(\frac{\partial}{\partial x_{j}} -\frac{\partial}{\partial x_{k}}\right)\psi \big|_{x_{j} = x_{k}^{-}} = g\psi \big|_{x_{j} = x_{k}}
$$

由于玻色子的交换对称性  

$$  
\left(\frac{\partial}{\partial x_{j}} -\frac{\partial}{\partial x_{k}}\right)\psi \big|_{x_{j} = x_{k}^{+}} = \left(\frac{\partial}{\partial x_{k}} -\frac{\partial}{\partial x_{j}}\right)\psi \big|_{x_{j} = x_{k}^{-}}
$$

可以得到同样的边界条件:  

$$  
\left(\frac{\partial}{\partial x_{j}} -\frac{\partial}{\partial x_{k}}\right)\psi \big|_{x_{j} = x_{k}^{+}} = c\psi \big|_{x_{j} = x_{k}^{-}}
$$

现在我们需要找出粒子不同排列时对应的系数关系, 因此, 我们取定区域 $x_{1}< x_{2}< \dots < x_{k}< x_{j}< \dots < x_{N}$ , 利用如下表示:  

$$  
\begin{array}{l}\phi (P) = A(P)\mathrm{e}^{\mathrm{i}(k p_{1}x_{1} + \dots +k p_{k}x_{k} + k p_{j}x_{j} + \dots +k p_{N}x_{N})}, \\ \phi (Q) = A(Q)\mathrm{e}^{\mathrm{i}(k p_{1}x_{1} + \dots +k p_{j}x_{k} + k p_{k}x_{j} + \dots +k p_{N}x_{N})}. \end{array}
$$

显然，整个区域上的波函数可写为 $\phi (x_{1},x_{2},\dots ,x_{N}) = [\dots +\phi (P) + \phi (Q) + \dots ]$，代入边界条件可得

$$
\left(\frac{\partial}{\partial x_{j}} -\frac{\partial}{\partial x_{k}}\right)[\dots +\phi (P) + \phi (Q) + \dots ]|_{x_{j} = x_{k}^{+}} = c[\dots +\phi (P) + \phi (Q) + \dots ]|_{x_{j} = x_{k}}.
$$

由此，可以进一步得到

$$
-{\mathrm{i}}(k_{p_{k}} - k_{p_{j}})[A(P) - A(Q)] = c[A(P) + A(Q)].
$$

因此，我们可以得到如下关系：

$$
A(Q) = -\frac{c - \mathrm{i}(k_{p_{j}} - k_{p_{k}})}{c + \mathrm{i}(k_{p_{j}} - k_{p_{k}})} A(P) = \frac{k_{p_{k}} - k_{p_{j}} - \mathrm{i}c}{k_{p_{k}} - k_{p_{j}} + \mathrm{i}c} A(P) = -\mathrm{e}^{-\mathrm{i}\theta (k_{p_{k}} - k_{p_{j}})} A(P).
$$

即

$$
A(k_{1},\dots ,k_{p_{j}},\dots ,k_{p_{k}},\dots ,k_{N}) = -\mathrm{e}^{-\mathrm{i}\theta (k_{p_{k}} - k_{p_{j}})} A(k_{1},\dots ,k_{p_{k}},\dots ,k_{p_{j}},\dots ,k_{N}),
$$

其中，$\theta (k_{i} - k_{j}) = 2\arctan \left(\frac{k_{i} - k_{j}}{c}\right)$。为了方便，我们取 $A(I) = 1$，这里 $A(I) = A(k_{1},\dots ,k_{N})$。

$$ {\color{red} 考虑波函数满足的周期性边界条件：} $$

$$
\psi (0,x_{2},\dots ,x_{N}) = \psi (x_{2},\dots ,x_{N},L).
$$

于是，可以得到

$$
A_{P}\mathrm{e}^{\mathrm{i}(k_{p_{2}}x_{2} + k_{p_{3}}x_{3} + \dots +k_{p_{N}}x_{N})} = A_{P^{\prime}}\mathrm{e}^{\mathrm{i}(k_{p_{2}}x_{2} + k_{p_{3}}x_{3} + \dots +k_{p_{N}}x_{N} + k_{p_{1}}L)}
$$

其中 $P = (\mathcal{P}_{1},\mathcal{P}_{2},\dots ,\mathcal{P}_{N})$ 和 $P^{'} = (\mathcal{P}_{2},\mathcal{P}_{3},\dots ,\mathcal{P}_{N},\mathcal{P}_{1})$，可得

$$
A_{P} = A_{P^{\prime}}\mathrm{e}^{\mathrm{i}k_{P_{1}}L}.
$$

继而（每次对换相邻两个）

$$
\begin{array}{r l}
& {A_{P_{1}P_{2}\dots P_{N}} = -\mathrm{e}^{\mathrm{i}\theta (k_{P_{1}} - k_{P_{2}})}A_{P_{2}P_{1}P_{3}\dots}}\\
& {\qquad = \left[-\mathrm{e}^{\mathrm{i}\theta (k_{P_{1}} - k_{P_{2}})}\right]\left[-\mathrm{e}^{\mathrm{i}\theta (k_{P_{1}} - k_{\mathcal{P}_{3}})}\right]A_{P_{2}P_{3}P_{1}P_{4}\dots}}\\
& {\qquad \vdots}\\
& {\qquad = (-1)^{N-1}\mathrm{e}^{\sum_{q}\mathrm{i}\theta (k_{P_{1}} - k_{q})}A_{P_{2}P_{3}\dots P_{N}P_{1}}}
\end{array}
$$

比较可得

$$
\mathrm{e}^{\mathrm{i}k_{P_{j}}L} = (-1)^{N-1}\mathrm{e}^{\sum_{q}\mathrm{i}\theta (k_{P_{j}} - k_{q})}.
$$
即

$$\textcolor{yellow}{
\mathrm{e}^{\mathrm{i}k_{j}L} = -\prod_{l = 1}^{N}\frac{k_{j} - k_{l} + \mathrm{i}c}{k_{j} - k_{l} - \mathrm{i}c},}
$$

我们称上式为 $\textcolor{yellow}{Bethe Ansatz 方程}$。通过求解它，我们便可以得到所有的准动量 $k_{j}(j = 1, \dots , N)$，从而给出体系的动量和能量。为此，我们先对上述 BA 方程两边取对数得到

$$\textcolor{yellow}{
k_{i}L = 2\pi I_{i} - \sum_{l = 1}^{N}2\arctan \frac{k_{i} - k_{l}}{c},}
$$

这里，$\{I_{i}\}$ 是准动量 $\{k_{i}\}$ 对应的量子数，其中

$$
I_{i} = \mathrm{整数}\qquad (N = \mathrm{奇数})，I_{i} + \frac{1}{2} = \mathrm{整数}\quad (N = \mathrm{偶数})
$$

对于基态，量子数 $\{I_{i}\}$ 的取值为 $\{I_{i}\} = \left\{-\frac{N - 1}{2}, -\frac{N - 3}{2}, \dots , \frac{N - 1}{2}\right\}$

$\textcolor{lime}{基态行为}$

一维 $\delta$ 相互作用的玻色气体，排斥相互作用情况 $(c > 0)$，取热力学极限: $N, L \to \infty$，粒子数密度 $n = N / L$ 为常数，则 BA 方程可由密度分布函数来表示。通过对 $N$ 个 BA 方程累乘的计算，我们得到一个关系: $\mathrm{e}^{\mathrm{i}\sum_{i}^{N}k_{i}L} = 1$，$ {\color{green} 因为上下抵消} ，$因此，总的动量为 $P = \sum_{j}^{N}k_{j} = 2\pi l / L(l = 0, \pm 1, \dots)$。

对于基态，$P = 0$，所有的 $k_{i}$ 在均匀区间 $[- q_0, q_0]$ 内对称分布，$q_0$ 为费米点。

我们引入区间 $[k, k + \Delta k]$ 内的准动量密度分布函数 $\rho (k) = \lim_{\Delta k \to 0} 1 / (L \Delta k)$，则此区间内的粒子数可表示为 $L\rho (k)\Delta k$。区间 $[I_{k}, I_{k} + \Delta I]$ 内的粒子数与区间 $[k, k + \Delta k]$ 内的粒子数相同 $\Delta I = L\rho (k)\Delta k$

我们得到
$$
k_{i}L = 2\pi I_{i} - \sum_{l = 1}^{N}2\arctan \frac{k_{i} - k_{l}}{c},
$$

$$
k = \frac{2\pi I}{L} -2\int_{-q_0}^{q_0}\arctan \frac{k - q}{c}\rho (q)\mathrm{d}q
$$

两边对k求导：
$$
1 = 2\pi \rho (k) - 2\int_{-q_0}^{q_0}\frac{c}{c^{2} + (k - q)^{2}}\rho (q)\mathrm{d}q
$$

换回国际单位制，$c=g/2$，补上$\frac{\hbar^{2}}{2m}$，其中$\tilde{g}=\frac{mg}{\hbar^2}$
$$
2\pi \rho (k) = 1 + 2\int_{-q_{0}}^{q_{0}}\frac{\tilde{g}\rho(k^{\prime})}{(k - k^{\prime})^{2} + \tilde{g}^{2}},
$$

这里 $q_{0}$ 是动量密度的边界，即对于任意 $|k| > q_{0}$，我们总有 $\rho (k) = 0$。在连续动量极限下，式也应被改写为

$$
E = L\int_{-q_{0}}^{q_{0}}\mathrm{d}k\frac{\hbar^{2}k^{2}}{2m}\rho (k),
$$

其中，原子数密度 $\rho_{0}$ 的定义为

$$
\rho_{0} = \int_{-q_{0}}^{q_{0}}\rho (k)\mathrm{d}k.
$$

在实际的科研中，我们通常将这些式子转化为无量纲的形式，最常用的方法为

$$
G(q) = \rho (k / q_{0}),\quad \alpha = \frac{\tilde{g}}{q_{0}},\quad \gamma = \frac{\tilde{g}}{\rho_{0}},
$$

运用上述方法将式和式去掉量纲，我们就可以得到$ {\color{yellow}Lieb-Liniger方程 } $，即

$$
\textcolor{yellow}{
\alpha = \gamma \int_{-1}^{+1}\mathrm{d}q G(q),}
$$

$$
\textcolor{yellow}{
G(q) = \frac{1}{2\pi} +\int_{-1}^{+1}\frac{dq^{\prime}}{2\pi} G(q^{\prime})\frac{2\alpha}{(q^{\prime} - q)^{2} + \alpha^{2}},}
$$

这里，$G$ 是我们拟设的无量纲态密度；$q$ 是无量纲动量；$\gamma$ 是控制系统相互作用大小的Lieb-Liniger参数

$$
e(\gamma) = \left( \frac{\gamma}{\alpha(\gamma)} \right)^3 \int_{-1}^{+1} dq \, G(q; \gamma) q^2.
$$

$$
\mu = \left. \frac{\partial E}{\partial N} \right|_L = \frac{\hbar^2}{2m} n^2 \left[ 3e(\gamma) - \gamma e'(\gamma) \right].
$$

给出系统的状态方程，即化学势与原子数密度的关系 $\mu = \mu(n)$。
